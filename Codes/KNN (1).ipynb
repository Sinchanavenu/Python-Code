```python
# Split the data into features and target
# X = data_cleaned.drop('Purchased', axis=1)
# y = data_cleaned['Purchased']
X=data.iloc[:,[2,3]].values

y=data.iloc[:,-1].values

from sklearn.model_selection import train_test_split
# Split into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Define the k-NN classifier
from sklearn.neighbors import KNeighborsClassifier
k = 5  # You can experiment with different values of k
knn = KNeighborsClassifier(n_neighbors=k)

# Train the model
knn.fit(X_train, y_train)

# Make predictions on the test set
y_pred = knn.predict(X_test)

# Evaluate the model
from sklearn import metrics
prediction = metrics.accuracy_score(y_test, y_pred)

classification_rep_knn = metrics.classification_report(y_test, y_pred)

print('Accuracy (k-NN):' , prediction *100)
print(classification_rep_knn)
```

```python
# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Initialize k-NN model with optimal k
optimal_k = 8
knn = KNeighborsClassifier(n_neighbors=optimal_k)
```

```python
# Train the model
knn.fit(X_train, y_train)
```

```python
# Predict on the test set
y_pred = knn.predict(X_test)
```

```python
# Calculate accuracy
accuracy = metrics.accuracy_score(y_test, y_pred)
print(f"Optimal k: {optimal_k}")
print(f"Accuracy: {accuracy * 100:.2f}%")
```
