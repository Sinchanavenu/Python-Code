```python
import requests
from bs4 import BeautifulSoup
url = "https://quotes.toscrape.com/"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")
quotes = soup.find_all("span", class_="text")
totalquotes=[]
for i in quotes:
  quote=i.text
  totalquotes.append(quote)
print(totalquotes)
```
