```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import numpy as np

# Load the Iris dataset
data = load_iris()
X = data.data[:, :2]  # Only the first two features (sepal length and sepal width)
y = data.target

# We'll use only classes 0 and 1 for simplicity, which are linearly separable
X = X[y != 2]
y = y[y != 2]

# Scatter plot of the dataset
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.coolwarm)
plt.title("Linear Dataset (Class 0 vs Class 1 from Iris)")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.show()


import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split

# Load Iris dataset (only using the first two features for simplicity)
data = load_iris()
X = data.data[:, :2]  # First two features: Sepal Length and Sepal Width
y = data.target

# We will only use classes 0 and 1 (Setosa and Versicolor) for linearly separable data
X = X[y != 2]
y = y[y != 2]

# Function to plot decision boundaries
def plot_decision_boundaries(X, y, model, ax):
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1), np.arange(y_min, y_max, 0.1))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    ax.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.coolwarm)
    ax.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', cmap=plt.cm.coolwarm)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Different kernels and combinations of C and gamma
kernels = ['linear', 'sigmoid', 'poly', 'rbf']
C_values = [0.1, 1, 10]
gamma_values = [0.1, 1, 'scale']

# Create subplots
fig, axes = plt.subplots(len(kernels), len(gamma_values), figsize=(15, 12))
fig.subplots_adjust(hspace=0.4, wspace=0.4)

# Loop over kernels and gamma values to train the models and plot decision boundaries
for i, kernel in enumerate(kernels):
    for j, gamma in enumerate(gamma_values):
        model = SVC(kernel=kernel, C=1, gamma=gamma)
        model.fit(X_train, y_train)
        ax = axes[i, j]
        plot_decision_boundaries(X, y, model, ax)
        ax.set_title(f"{kernel.capitalize()} Kernel\nC=1, gamma={gamma}")

plt.show()

```
