---
How to find or validate an email address

---

```python
email_pattern = r"^[A-Z0-9._%+-]+@[A-Z0-9.-]+\.[A-Z]{2,4}$"

user = input("Enter an email address: ")

if re.match(email_pattern, user, re.IGNORECASE):
    print(f"'{user}' matches the pattern")
else:
    print(f"'{user}' does NOT match the pattern")
```

---
Matching a Valid Date

---

```python
date_pattern = r"^(0[1-9]|1[012])[- /.](0[1-9]|[12][0-9]|3[01])[- /.](19|20)\d\d$"

user = input("Enter a date (MM/DD/YYYY): ")

if re.match(date_pattern, user):
    print(f"'{user}' matches the pattern")
else:
    print(f"'{user}' does NOT match the pattern")
```

---
Match an American Express Credit Card Number which always begin with 34 or 37 and
totals 15 digits.

---

```python
cc_pattern = r"^3[47]\d{13}$"

user = input("Enter an American Express card number (15 digits, starts with 34 or 37): ")

if re.match(cc_pattern, user):
    print(f"'{user}' matches the pattern")
else:
    print(f"'{user}' does NOT match the pattern")
```

---
Match a full U.S. Phone Number: +1-(555)-555-5555

---

```python
phone_pattern = r"^\+1-\(\d{3}\)-\d{3}-\d{4}$"

user = input("Enter a U.S. phone number (+1-(555)-555-5555): ")

if re.match(phone_pattern, user):
    print(f"'{user}' matches the pattern")
else:
    print(f"'{user}' does NOT match the pattern")
```
