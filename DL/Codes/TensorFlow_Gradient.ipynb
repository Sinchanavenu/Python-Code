```python
import numpy as np
```

```python
I = np.identity(3)
print(I)
A = np.reshape(np.arange(24)+1, (8,3))
print(A)
output_gradient = np.random.rand(8,3)
AA = (I - A[:, :, np.newaxis]) * A[:, np.newaxis , :]
# Compute g^(0)T * (I - a^(0)) ⊙ a^(0)T using einsum
result = np.einsum('ij,ijk->ik', output_gradient, AA)

print("\nResult of g^(0)T * (I - a^(0)) ⊙ a^(0)T using einsum:")
print(result)
```
