```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split

# Generate circular dataset with a larger sample size
X, y = make_circles(n_samples=3000, noise=0.05, factor=0.5, random_state=42)  # Change n_samples as needed (1000-5000)

# Function to plot decision boundaries
def plot_decision_boundaries(X, y, model, ax):
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1), np.arange(y_min, y_max, 0.1))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    ax.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.coolwarm)
    ax.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', cmap=plt.cm.coolwarm)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Different kernels and combinations of C and gamma
kernels = ['linear', 'sigmoid', 'poly', 'rbf']
C_values = [1]  # Using a fixed C value for simplicity
gamma_values = [0.1, 1, 'scale']

# Create subplots
fig, axes = plt.subplots(len(kernels), len(gamma_values), figsize=(15, 12))
fig.subplots_adjust(hspace=0.4, wspace=0.4)

# Loop over kernels and gamma values to train the models and plot decision boundaries
for i, kernel in enumerate(kernels):
    for j, gamma in enumerate(gamma_values):
        model = SVC(kernel=kernel, C=1, gamma=gamma)
        model.fit(X_train, y_train)
        ax = axes[i, j]
        plot_decision_boundaries(X, y, model, ax)
        ax.set_title(f"{kernel.capitalize()} Kernel\nC=1, gamma={gamma}")

plt.show()
```
