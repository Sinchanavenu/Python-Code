```python
import requests
from bs4 import BeautifulSoup
import csv

# URL for scraping data
url = 'http://books.toscrape.com/'

# Get the page's HTML content
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

data = []
books =soup.find_all('article', class_='product_pod')
# Extract book information
for book in books:
    title = book.h3.a['title']
    price = book.find('p', class_='price_color').text
    availability = book.find('p', class_='availability').text.strip()

    data.append((title, price, availability))



# Define the CSV file path
csv_file_path = '241057007SinchanaVenugopal.csv'

# Write data to CSV file
with open(csv_file_path, 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    # Write the header row
    writer.writerow(['Title', 'Price', 'Availability'])
    # Write the data rows
    writer.writerows(data)

print("Data successfully saved to 241057007_SinchanaVenugopal.csv ")
```

```python
#without csv file
import requests
from bs4 import BeautifulSoup
import texttable as tt


url = 'http://books.toscrape.com/'


page = requests.get(url)
soup = BeautifulSoup(page.content, 'html.parser')

data = []


for book in soup.find_all('article', class_='product_pod'):
    title = book.h3.a['title']
    price = book.find('p', class_='price_color').text
    availability = book.find('p', class_='availability').text.strip()

    data.append((title, price, availability))


table = tt.Texttable()
table.add_rows([['Title', 'Price', 'Availability']] + data)
table.set_cols_align(['c', 'r', 'l'])
print(table.draw())
```
