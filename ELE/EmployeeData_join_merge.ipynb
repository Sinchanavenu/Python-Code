```python
import numpy as np
import pandas as pd
```

```python
employee_data = {
    'EmployeeID': [1, 2, 3, 4, 5, 6, 7, 8],
    'Education': ['Bachelors', 'Bachelors', 'Bachelors', 'Masters', np.nan, 'PhD', 'PhD', 'Masters'],
    'JoiningYear': [2017, 2013, 2014, 2016, 2017, np.nan, 2018, 2019],
    'City': ['Bangalore', 'Pune', np.nan, 'Bangalore', 'Pune', 'New Delhi', 'Bangalore', 'Pune'],
    'PaymentTier': [3, 1, 3, np.nan, 3, 2, 1, 3],
    'Age': [34, 28, 38, 27, 24, 32, np.nan, 30],
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Male', 'Female', 'Male', 'Female'],
    'EverBenched': ['No', 'No', 'No', 'No', 'Yes', 'Yes', 'No', 'Yes'],
    'ExperienceInCurrentDomain': [0, 3, 2, 5, 2, 4, 1, np.nan],
    'LeaveOrNot': [0, 1, 0, 1, 1, 0, 1, 0],
    'Salary': [70000, 55000, 60000, 80000, 75000, 85000, 90000, np.nan]
}
```

```python
# Convert dictionary to DataFrame
employee_data = pd.DataFrame(employee_data)
employee_data
```

```python
# Setting hierarchical index using 'Gender' and 'City'
employee_data_hierarchical = employee_data.set_index(['Gender', 'City'])
employee_data_hierarchical
```

**Hierarchical Indexing**

```python
# Access data for all 'Male' employees in 'Bangalore'
male_bangalore_data = employee_data_hierarchical.loc[('Male', 'Bangalore')]
print(male_bangalore_data)

# Access 'Education' column data for all 'Female' employees in 'Pune'
female_pune_education = employee_data_hierarchical.loc[('Female', 'Pune'), 'Education']
print(female_pune_education)
```

```python
# Access data in the first row (position-based) which corresponds to the first 'Male', 'Bangalore'
first_row_data = employee_data_hierarchical.iloc[0]
print(first_row_data)

# Access data for the first row and first column (position-based)
first_row_first_col = employee_data_hierarchical.iloc[0, 0]
print(first_row_first_col)
```

```python

# Creating hierarchical index for rows (Gender, City)
row_index = [employee_data['Gender'], employee_data['City']]

# Creating multi-level (hierarchical) columns
columns = pd.MultiIndex.from_tuples([
    ('Education', 'Details'),
    ('Salary', 'Details'),
    ('ExperienceInCurrentDomain', 'Details'),
    ('Age', 'Details')
])

# Constructing hierarchical DataFrame with unique indices
employee_data_hierarchical = pd.DataFrame(
    employee_data[['Education', 'Salary', 'ExperienceInCurrentDomain', 'Age']].values,
    index=employee_data_hierarchical.index,
    columns=columns
)


# Set the names for the levels in the column index
employee_data_hierarchical.columns.names = ['Attribute', 'Details']

# Display the DataFrame
print(employee_data_hierarchical)
```

**Sort and swap**

```python
# Swap levels in the row index (Gender and City)
swapped_employee_data = employee_data_hierarchical.swaplevel(0, 1)

# Sort by the new row index after the swap
sorted_employee_data = swapped_employee_data.sort_index()

# Display the DataFrame after swapping and sorting
print(sorted_employee_data)
```

```python
# Remove duplicate rows based on all columns
employee_data_cleaned = employee_data.drop_duplicates()

# Create hierarchical index for rows (Gender, City)
employee_data_hierarchical = employee_data_cleaned.set_index(['Gender', 'City'])

# Creating multi-level (hierarchical) columns
columns = pd.MultiIndex.from_tuples([
    ('Education', 'Details'),
    ('Salary', 'Details'),
    ('ExperienceInCurrentDomain', 'Details'),
    ('Age', 'Details')
])

# Constructing hierarchical DataFrame
employee_data_hierarchical = pd.DataFrame(
    employee_data_cleaned[['Education', 'Salary', 'ExperienceInCurrentDomain', 'Age']].values,
    index=employee_data_hierarchical.index,
    columns=columns
)

# Set the names for the levels in the column index
employee_data_hierarchical.columns.names = ['Attribute', 'Details']

# Display the hierarchical DataFrame
print("Hierarchical DataFrame:\n", employee_data_hierarchical)

```

```python
# Check for duplicates in the index and aggregate if necessary
if employee_data_hierarchical.index.duplicated().any():
    employee_data_hierarchical = employee_data_hierarchical.groupby(level=[0, 1]).first()

# Display the cleaned hierarchical DataFrame
print("\nHierarchical DataFrame after handling duplicates:\n", employee_data_hierarchical)

```

**Stack and Unstack**

```python
# Apply stack() to the hierarchical columns
stacked_data = employee_data_hierarchical.stack()

# Display the stacked DataFrame
print("\nStacked DataFrame:\n", stacked_data)

```

```python
# Apply unstack() to the hierarchical rows
unstacked_data = stacked_data.unstack()

# Display the unstacked DataFrame
print("\nUnstacked DataFrame:\n", unstacked_data)
```

**Merge**

```python
# Sample data for merging
salary_data_dict = {
    'EmployeeID': [1, 2, 3, 4, 5, 6, 8],
    'Bonus': [5000, 2000, 3000, 4000, 6000, 7000, 1500]
}

# Create another DataFrame for merging
salary_data = pd.DataFrame(salary_data_dict)

# Display the original DataFrames
print("Employee Data:\n", employee_data)
print("\nSalary Data:\n", salary_data)
```

```python
# Merge using default inner join on EmployeeID
merged_data_inner = employee_data.merge(salary_data, on='EmployeeID', how='inner')

# Display the merged DataFrame
print("\nMerged Data (Inner Join):\n", merged_data_inner)
```

```python
# Merge using outer join
merged_data_outer = employee_data.merge(salary_data, on='EmployeeID', how='outer')

# Display the merged DataFrame
print("\nMerged Data (Outer Join):\n", merged_data_outer)
```

```python
# Merge using left join
merged_data_left = employee_data.merge(salary_data, on='EmployeeID', how='left')

# Display the merged DataFrame
print("\nMerged Data (Left Join):\n", merged_data_left)
```

```python
# Merge using right join
merged_data_right = employee_data.merge(salary_data, on='EmployeeID', how='right')

# Display the merged DataFrame
print("\nMerged Data (Right Join):\n", merged_data_right)
```

```python
# Set EmployeeID as index for both DataFrames
employee_data_indexed = employee_data.set_index('EmployeeID')
salary_data_indexed = salary_data.set_index('EmployeeID')

# Join using the index
joined_data = employee_data_indexed.join(salary_data_indexed, how='inner', lsuffix='_emp', rsuffix='_sal')

# Display the joined DataFrame
print("\nJoined Data:\n", joined_data)
```

**Concat**

```python
# Concatenate the DataFrames along rows
concatenated_data = pd.concat([employee_data, salary_data], axis=0, ignore_index=True)

# Display the concatenated DataFrame
print("\nConcatenated Data:\n", concatenated_data)
```

**Reshape and pivot**

```python
# Pivot the DataFrame to show average salary by education and city
pivot_data = employee_data.pivot(index='Education', columns='City', values='Salary')

# Display the pivoted DataFrame
print("\nPivoted Data:\n", pivot_data)
```

```python
# Create a pivot table to show average salary by education and joining year
pivot_table_data = employee_data.pivot_table(index='Education', columns='JoiningYear', values='Salary', aggfunc='mean')

# Display the pivot table
print("\nPivot Table Data:\n", pivot_table_data)
```
